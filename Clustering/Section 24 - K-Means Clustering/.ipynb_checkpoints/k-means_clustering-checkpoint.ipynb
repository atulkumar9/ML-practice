# K-MEANS CLUSTERING

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

Spending score is the score that mall computed for each client depending on the features like Anual Income, The no of times client appear in the mall the amount they spend in year, So spending score is a matrix of values between [1, 100]. 
The closer the spending score to 100 more the client spends and vica-versa.
Now we need to segment the clients into different groups based on anual income and spending score, so we need a clustring algorithm because we dont know anything about the size of clusters and no of clusters we need to take into concederation

```python
X = dataset.iloc[:, 3:5].values
```

```python
#Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    #within clusters sum of square
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

So we can see 5 is the elbow point and so it is the optimal number of clusters to be taken.

```python
#Applying k-means to the mall dataset
kmeans = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
#Visualising the clusters(only for 2-D)
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', marker = 'o',  label = 'Careful')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'green', marker = 'v', label = 'Standard')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'blue', marker = 'p', label = 'Target')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', marker = 'd', label = 'Careless')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', marker = '+', label = 'Sensible')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'black', marker = '*', label = 'centroids')
plt.title('clusters of clients')
plt.xlabel('Anual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import DBSCAN 
db = DBSCAN(eps = 9, min_samples = 5, metric = 'euclidean')
y_db = db.fit_predict(X)
y_db
```

```python
plt.scatter(X[y_db == 0, 0], X[y_db == 0,1], c = 'lightblue', marker = 'o', s = 40, label = 'cluster 1')
plt.scatter(X[y_db == -1, 0], X[y_db == -1,1], c = 'red', marker = 'o', s = 40, label = 'cluster 2')
plt.scatter(X[y_db == 1, 0], X[y_db == 1,1], c = 'green', marker = 'o', s = 40, label = 'cluster 3')
plt.scatter(X[y_db == 2, 0], X[y_db == 2,1], c = 'yellow', marker = 'o', s = 40, label = 'cluster 4')
plt.scatter(X[y_db == 3, 0], X[y_db == 3,1], c = 'pink', marker = 'o', s = 40, label = 'cluster 5')
plt.scatter(X[y_db == 4, 0], X[y_db == 4,1], c = 'black', marker = 'o', s = 40, label = 'cluster 6')
plt.scatter(X[y_db == 5, 0], X[y_db == 5,1], c = 'cyan', marker = 'o', s = 40, label = 'cluster 7')
plt.scatter(X[y_db == 6, 0], X[y_db == 6,1], c = 'orange', marker = 'o', s = 40, label = 'cluster 8')
plt.scatter(X[y_db == 7, 0], X[y_db == 7,1], c = 'violet', marker = 'o', s = 40, label = 'cluster 9')
plt.legend()
plt.show()
```

So we can see DBCSAN Clustering model is not good enough for the problem dataset, hence best option is K-Mean and next best is Hierarchical clustering.
